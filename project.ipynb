```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pydicom

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
import os
import cv2
import numpy as np

dataset_path = r"C:\Users\NDC\Desktop\dataset"

images = []
labels = []

IMG_SIZE = 128

classes = {
    "Bengin cases": 0,
    "Malignant cases": 1
}

for folder_name, label in classes.items():

    folder = os.path.join(dataset_path, folder_name)

    for file in os.listdir(folder):

        path = os.path.join(folder, file)

        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

        if img is not None:

            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))   # IMPORTANT FIX

            images.append(img)
            labels.append(label)

images = np.array(images)
labels = np.array(labels)

print("Images shape:", images.shape)
print("Labels shape:", labels.shape)
```

```python
images = images.reshape(-1,128,128,1)

print(images.shape)
```

```python
images = images / 255.0
```

```python
segmented_images = []

for img in images:
    
    img2 = img.reshape(128,128)   # remove channel temporarily
    
    _, thresh = cv2.threshold((img2*255).astype(np.uint8), 127, 255, cv2.THRESH_BINARY_INV)
    
    kernel = np.ones((5,5),np.uint8)
    
    mask = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    
    mask = mask / 255.0
    
    segmented = img2 * mask
    
    segmented = segmented.reshape(128,128,1)
    
    segmented_images.append(segmented)

segmented_images = np.array(segmented_images)

print(segmented_images.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    segmented_images,
    labels,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
model = models.Sequential()

model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(128,128,1)))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D(2,2))

model.add(layers.Flatten())

model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dropout(0.5))

model.add(layers.Dense(1,activation='sigmoid'))

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=16,
    validation_split=0.1
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=16,
    validation_split=0.1
)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.legend(["Train","Validation"])
plt.show()
```

```python
pred = model.predict(X_test)

pred = (pred > 0.5).astype(int)

print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
sample = X_test[6]

plt.imshow(sample.reshape(128,128),cmap='gray')
plt.title("CT Image")
plt.show()

prediction = model.predict(sample.reshape(1,128,128,1))

if prediction > 0.5:
    print("Malignant Lung Cancer Detected")
else:
    print("Benign Lung Case")
```
